```python
import pandas as pd
```

```python
df = pd.read_csv("/content/News Category (Text).csv")
```

```python
df
```

```python
X = df['headline']
```

```python
Y = df['category']
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
```

```python
X_encode = CountVectorizer().fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X_encode,Y, random_state=112)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
```

```python
model = LogisticRegression(random_state=112)
model.fit(X_train,Y_train)
pred = model.predict(X_test)
print(accuracy_score(Y_test,pred))
print(confusion_matrix(Y_test,pred))
```

```python
model = RandomForestClassifier(random_state=112)
model.fit(X_train,Y_train)
pred = model.predict(X_test)
print(accuracy_score(Y_test,pred))
print(confusion_matrix(Y_test,pred))
```
