```python
from groq import Groq
```

```python
import os
from google.colab import userdata

os.environ['GROQ_API_KEY'] = userdata.get('GROQ_API_KEY')
client = Groq()
```

```python
q = input("query : ")
```

```python
r = client.chat.completions.create(
    model = "llama-3.1-8b-instant",
    messages = [
        {"role": "user", "content": q}
    ]
)
```

```python
r.choices[0].message.content
```
